```python
file0 = "/content/mitochondrial_genomes.fasta"
file1 = "/content/sequence (1).fasta"  # Замените на точное имя первого файла
file2 = "/content/sequence.fasta"  # Замените на точное имя второго файла
combined_file = "combined_sequences.fasta"
```

```python
# Чтение и объединение последовательностей из двух файлов
records = []
for file in [file0, file1, file2]:
    records.extend(list(SeqIO.parse(file, "fasta")))

# Сохранение объединенных последовательностей в новый файл
with open(combined_file, "w") as output_handle:
    SeqIO.write(records, output_handle, "fasta")

print(f"Объединенные последовательности сохранены в {combined_file}")
```

```python
from Bio import SeqIO

# Загрузка последовательностей
sequences = list(SeqIO.parse("/content/complete_genome.fasta", "fasta"))

# Фильтрация
filtered_sequences = []
for seq in sequences:
    if len(seq.seq) > 500 and 'N' not in seq.seq.upper():
        filtered_sequences.append(seq)

# Сохранение отфильтрованных последовательностей
SeqIO.write(filtered_sequences, "18S_rRNA_filtered_sequences.fasta", "fasta")
print(f"После фильтрации осталось {len(filtered_sequences)} последовательностей.")
```
